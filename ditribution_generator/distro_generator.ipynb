# Exercise: Generate different distributions
Use the uniform distribution to generate an exponential distribution $p(x) = e^{-x}$ and a Gaussian distribution (using the formular in the lecture). Plot the histogram to check the results.

# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Generators definitions

```python
class ExponentialGenerator:
    def generate(self, n_points):
        r = np.random.uniform(low=0, high=1, size=n_points)

        transform_func = lambda x: - np.log(1-x)

        return transform_func(r)
```

Explonation:
> According to the lecture, first we have to compute the distribuant of $e^{-x}$, then we can create following equation $r=1−e^{-x}$, after calculating we have $x=−ln(1−r)$. That all r is out numbers from uniform distribution...

```python
class GaussianGenerator:
    def generate(self, n_points):
        # they have to be uncorrelated
        r = np.random.uniform(low=0, high=1, size=n_points)
        s = np.random.uniform(low=0, high=1, size=n_points)

        transform_func = lambda x,y: np.sqrt(-2 * np.log(x)) * np.cos(2 * np.pi * y)

        return transform_func(r, s)
```

Explonation:
> Just took the formulas from the lecture...

# Testing

## Exponential

```python
generator = ExponentialGenerator()

x_pred = generator.generate(100000)

plt.hist(x_pred, bins=100, density=True, alpha=0.7, label='Simulation')

# theoretic perfect model
t = np.linspace(0, 7, 100)
plt.plot(t, np.exp(-t), 'r', label='Theory')
plt.legend()
plt.show()
```

## Gaussian

```python
generator = GaussianGenerator()

x_pred = generator.generate(100000)

plt.hist(x_pred, bins=100, density=True, alpha=0.7, color='green', label='Simulation')

# theoretic perfect model
t = np.linspace(-4, 4, 100)
plt.plot(t, 1/(np.sqrt(2*np.pi)) * np.exp(-0.5*t**2), 'r', label='Theory')
plt.title("Gauss distribution")
plt.legend()
plt.show()
```
